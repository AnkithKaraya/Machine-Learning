# Face Detection Using Haar Cascades and MTCNN

## Haar Cascades

```python
# Importing the opencv library
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Read the image and convert into gray image
image = cv.imread('Image_1.jpg')
gray_img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img)
```

```python
haar = cv.CascadeClassifier('C:/Users/LENOVO/Dropbox/PC/Desktop/Machine_Learning_Project/haarcascade.xml')

face_det = haar.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)
```

```python
print("No. of faces detected: ", len(face_det))
```

```python
# Draw a rectangle over the faces detected

for (x, y, w, h) in face_det:
    cv.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), thickness=2)
    
plt.imshow(image)
```

## MTCNN

```python
import mtcnn.mtcnn
```

```python
img = cv.imread('Image_14.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
det = mtcnn.MTCNN()

Faces_detected = det.detect_faces(img)
```

```python
print("Number of faces detected: ", len(Faces_detected))
```

```python
# Drawing rectangle on the faces

for i in range(len(Faces_detected)):
    a, b, c, d = Faces_detected[i]['box']
    cv.rectangle(img, (a,b), (a+c, b+d), (255,0,0), thickness=2)

    plt.imshow(img)
```
